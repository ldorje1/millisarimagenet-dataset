### <font color="orange"> (1) Core SquiggleMilli Model
Define models (copied from main_quantifier.py) This is their exact architecture, just without the giant training loop at the bottom and without tf.compat.v1.enable_eager_execution().

```python
# -------- Discriminator --------
def Discriminator():
    # 3D branch
    input_layer_3D = Input((32, 64, 96, 1), name='input_layer_3D')
    x3 = Conv3D(16, (6,6,6), strides=(2,2,2), padding='same', name='conv_layer1_3D')(input_layer_3D)
    x3 = BatchNormalization(name='bnorm_layer1_3D')(x3)
    x3 = LeakyReLU(name='lrelu_layer1_3D')(x3)
    x3 = GaussianNoise(0.5)(x3)

    x3 = Conv3D(32, (6,6,6), strides=(2,2,2), padding='same', name='conv_layer2_3D')(x3)
    x3 = BatchNormalization(name='bnorm_layer2_3D')(x3)
    x3 = LeakyReLU(name='lrelu_layer2_3D')(x3)
    x3 = GaussianNoise(0.5)(x3)

    x3 = Conv3D(64, (6,6,6), strides=(2,2,2), padding='same', name='conv_layer3_3D')(x3)
    x3 = BatchNormalization(name='bnorm_layer3_3D')(x3)
    x3 = LeakyReLU(name='lrelu_layer3_3D')(x3)
    x3 = GaussianNoise(0.5)(x3)

    x3 = Conv3D(128, (6,6,6), strides=(2,2,2), padding='same', name='conv_layer4_3D')(x3)
    x3 = BatchNormalization(name='bnorm_layer4_3D')(x3)
    x3 = LeakyReLU(name='lrelu_layer4_3D')(x3)

    x3 = Conv3D(256, (6,6,6), strides=(2,2,2), padding='same', name='conv_layer5_3D')(x3)
    x3 = BatchNormalization(name='bnorm_layer5_3D')(x3)
    x3 = LeakyReLU(name='lrelu_layer5_3D')(x3)

    x3 = Conv3D(1024, (6,6,6), strides=(2,2,2), padding='same', name='conv_layer6_3D')(x3)
    x3 = BatchNormalization(name='bnorm_layer6_3D')(x3)
    x3 = LeakyReLU(name='lrelu_layer6_3D')(x3)
    x3 = Flatten(name='flatten_layer_3D')(x3)
    x3 = Dense(512, name='dense_layer_3D')(x3)
    x3 = Reshape((1,1,512), name='reshape_layer1_3D')(x3)

    # 2D branch
    input_layer_2D = Input((128, 256, 1), name='input_layer_2D_dis')
    x2 = Conv2D(4,  (4,3), strides=(2,2), padding='same', name='conv_layer1_2D')(input_layer_2D)
    x2 = BatchNormalization(name='bnorm_layer1_2D')(x2)
    x2 = LeakyReLU(name='lrelu_layer1_2D')(x2)

    x2 = Conv2D(8,  (6,6), strides=(2,2), padding='same', name='conv_layer2_2D')(x2)
    x2 = BatchNormalization(name='bnorm_layer2_2D')(x2)
    x2 = LeakyReLU(name='lrelu_layer2_2D')(x2)

    x2 = Conv2D(16, (6,6), strides=(2,2), padding='same', name='conv_layer3_2D')(x2)
    x2 = BatchNormalization(name='bnorm_layer3_2D')(x2)
    x2 = LeakyReLU(name='lrelu_layer3_2D')(x2)

    x2 = Conv2D(32, (6,6), strides=(2,2), padding='same', name='conv_layer4_2D')(x2)
    x2 = BatchNormalization(name='bnorm_layer4_2D')(x2)
    x2 = LeakyReLU(name='lrelu_layer4_2D')(x2)

    x2 = Conv2D(64, (6,6), strides=(2,2), padding='same', name='conv_layer5_2D')(x2)
    x2 = BatchNormalization(name='bnorm_layer5_2D')(x2)
    x2 = LeakyReLU(name='lrelu_layer5_2D')(x2)

    x2 = Conv2D(128,(6,6), strides=(2,2), padding='same', name='conv_layer6_2D')(x2)
    x2 = BatchNormalization(name='bnorm_layer6_2D')(x2)
    x2 = LeakyReLU(name='lrelu_layer6_2D')(x2)

    x2 = Conv2D(256,(6,6), strides=(2,2), padding='same', name='conv_layer7_2D')(x2)
    x2 = BatchNormalization(name='bnorm_layer7_2D')(x2)
    x2 = LeakyReLU(name='lrelu_layer7_2D')(x2)
    x2 = Reshape((1,1,512), name='reshape_layer1_2D_dis')(x2)

    # concat + head
    x = Concatenate(axis=3)([x3, x2])
    x = Flatten(name='flatten_layer_dis')(x)
    x = Dense(256, activation='relu', name='dense_layer_dis1')(x)
    out = Dense(1, activation='sigmoid', name='output_layer_dis')(x)

    model = Model(inputs=[input_layer_3D, input_layer_2D], outputs=out, name='Discriminator')
    opt = Adam(learning_rate=LR_D, beta_1=0.5)
    model.compile(loss='binary_crossentropy', optimizer=opt, metrics=['accuracy'])
    return model


# -------- MetricsComputation (5-D regression head) --------
def MetricsComputation():
    inp = Input((128, 256, 1), name='input_layer_2D_metrics')
    x = Conv2D(4,  (4,3), strides=(2,2), padding='same', name='conv_layer1_2D_m')(inp)
    x = BatchNormalization(name='bnorm_layer1_2D_m')(x)
    x = LeakyReLU(name='lrelu_layer1_2D_m')(x)

    x = Conv2D(8,  (6,6), strides=(2,2), padding='same', name='conv_layer2_2D_m')(x)
    x = BatchNormalization(name='bnorm_layer2_2D_m')(x)
    x = LeakyReLU(name='lrelu_layer2_2D_m')(x)

    x = Conv2D(16, (6,6), strides=(2,2), padding='same', name='conv_layer3_2D_m')(x)
    x = BatchNormalization(name='bnorm_layer3_2D_m')(x)
    x = LeakyReLU(name='lrelu_layer3_2D_m')(x)

    x = Conv2D(32, (6,6), strides=(2,2), padding='same', name='conv_layer4_2D_m')(x)
    x = BatchNormalization(name='bnorm_layer4_2D_m')(x)
    x = LeakyReLU(name='lrelu_layer4_2D_m')(x)

    x = Conv2D(64, (6,6), strides=(2,2), padding='same', name='conv_layer5_2D_m')(x)
    x = BatchNormalization(name='bnorm_layer5_2D_m')(x)
    x = LeakyReLU(name='lrelu_layer5_2D_m')(x)

    x = Conv2D(128,(6,6), strides=(2,2), padding='same', name='conv_layer6_2D_m')(x)
    x = BatchNormalization(name='bnorm_layer6_2D_m')(x)
    x = LeakyReLU(name='lrelu_layer6_2D_m')(x)

    x = Conv2D(256,(6,6), strides=(2,2), padding='same', name='conv_layer7_2D_m')(x)
    x = BatchNormalization(name='bnorm_layer7_2D_m')(x)
    x = LeakyReLU(name='lrelu_layer7_2D_m')(x)

    x = Reshape((1,1,512), name='reshape_layer1_2D_m')(x)
    x = Flatten(name='flatten_layer_dis_m')(x)
    x = Dense(256, activation='relu', name='dense_layer_dis1_m')(x)
    x = Dense(256, activation='relu', name='dense_layer_dis2_m')(x)
    x = Dense(256, activation='relu', name='dense_layer_dis3_m')(x)
    x = Dense(256, activation='relu', name='dense_layer_dis4_m')(x)
    out = Dense(5, name='output_layer_dis_m')(x)

    model = Model(inputs=inp, outputs=out, name='Metrics')
    return model


# -------- Generator (3D→2D with skip) --------
def Generator():
    input_layer = Input((32, 64, 96, 1), name='input_layer')
    skip_input  = Input((32, 64, 8),      name='skiplayer')

    x = Conv3D(16, (6,6,6), strides=(2,2,2), padding='same', name='conv_layer1')(input_layer)
    x = BatchNormalization(name='bnorm_layer1')(x)
    x = LeakyReLU(name='lrelu_layer1')(x)

    x = Conv3D(32, (6,6,6), strides=(2,2,2), padding='same', name='conv_layer2')(x)
    x = BatchNormalization(name='bnorm_layer2')(x)
    x = LeakyReLU(name='lrelu_layer2')(x)

    x = Conv3D(64, (6,6,6), strides=(2,2,2), padding='same', name='conv_layer3')(x)
    x = BatchNormalization(name='bnorm_layer3')(x)
    x = LeakyReLU(name='lrelu_layer3')(x)

    x = Conv3D(128,(6,6,6), strides=(2,2,2), padding='same', name='conv_layer4')(x)
    x = BatchNormalization(name='bnorm_layer4')(x)
    x = LeakyReLU(name='lrelu_layer4')(x)

    x = Conv3D(256,(6,6,6), strides=(2,2,2), padding='same', name='conv_layer5')(x)
    x = BatchNormalization(name='bnorm_layer5')(x)
    x = LeakyReLU(name='lrelu_layer5')(x)

    x = Conv3D(1024,(6,6,6), strides=(2,2,2), padding='same', name='conv_layer6')(x)
    x = BatchNormalization(name='bnorm_layer6')(x)
    x = LeakyReLU(name='lrelu_layer6')(x)

    x = Reshape((1,1,2048), name='reshape_layer1')(x)

    x = Conv2DTranspose(1024, (4,3), strides=(1,2), padding='same', name='deconv_layer1')(x)
    x = BatchNormalization(name='bnorm_layer7')(x)
    x = ReLU(name='relu_layer1')(x)

    x = Conv2DTranspose(512, (4,4), strides=(2,2), padding='same', name='deconv_layer2')(x)
    x = BatchNormalization(name='bnorm_layer8')(x)
    x = ReLU(name='relu_layer2')(x)

    x = Conv2DTranspose(256, (4,4), strides=(2,2), padding='same', name='deconv_layer3')(x)
    x = BatchNormalization(name='bnorm_layer9')(x)
    x = ReLU(name='relu_layer3')(x)

    x = Conv2DTranspose(128, (4,4), strides=(2,2), padding='same', name='deconv_layer4')(x)
    x = BatchNormalization(name='bnorm_layer10')(x)
    x = ReLU(name='relu_layer4')(x)

    x = Conv2DTranspose(64, (4,4), strides=(2,2), padding='same', name='deconv_layer5')(x)
    x = BatchNormalization(name='bnorm_layer11')(x)
    x = ReLU(name='relu_layer5')(x)

    x = Conv2DTranspose(16, (4,4), strides=(2,2), padding='same', name='deconv_layer6')(x)
    x = BatchNormalization(name='bnorm_layer12')(x)
    x = ReLU(name='relu_layer6')(x)

    x = Concatenate(axis=3)([x, skip_input])

    x = Conv2DTranspose(8, (4,4), strides=(2,2), padding='same', name='deconv_layer7')(x)
    x = BatchNormalization(name='bnorm_layer13')(x)
    x = ReLU(name='relu_layer7')(x)

    out = Conv2DTranspose(1, (4,4), strides=(2,2), padding='same', name='deconv_layer8')(x)

    model = Model(inputs=[input_layer, skip_input], outputs=out, name='Generator')
    return model


# -------- Combined GAN model (G + D + Metrics) --------
'''
def build_gan(g_model, d_model, m_model):
    d_model.trainable = False
    mm_in, skip_in = g_model.input
    gen_out = g_model.output             # (B,128,256,1)
    gan_out = d_model([mm_in, gen_out])  # real/fake
    m_out   = m_model(gen_out)           # 5-dim

    gen_out_rgb = K.concatenate([gen_out, gen_out, gen_out], axis=-1)

    gan_model = Model([mm_in, skip_in], [gen_out_rgb, gan_out, m_out], name='GAN')
    opt = Adam(learning_rate=LR_GAN, beta_1=0.5)
    gan_model.compile(
        loss=[tf.keras.losses.MeanAbsoluteError(),
              tf.keras.losses.BinaryCrossentropy(),
              tf.keras.losses.MeanAbsoluteError()],
        loss_weights=[LAMBDA_I, LAMBDA_BCE, LAMBDA_F],
        optimizer=opt
    )
    return gan_model
  '''
from tensorflow.keras.layers import Concatenate
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
import tensorflow as tf

def build_gan(g_model, d_model, m_model):
    # freeze discriminator inside the GAN
    d_model.trainable = False

    # KerasTensors
    mm_in, skip_in = g_model.inputs          # (B,32,64,96,1), (B,32,64,8)
    gen_out        = g_model.outputs[0]      # (B,128,256,1)

    # Discriminator output: real/fake
    gan_out = d_model([mm_in, gen_out])

    # Metrics head: 5-D features
    m_out = m_model(gen_out)

    # Make 3-channel image via Keras Concatenate layer
    gen_out_rgb = Concatenate(axis=-1, name="gen_out_rgb")(
        [gen_out, gen_out, gen_out]
    )

    # Combined model
    gan_model = Model(
        inputs=[mm_in, skip_in],
        outputs=[gen_out_rgb, gan_out, m_out],
        name='GAN'
    )

    opt = Adam(learning_rate=LR_GAN, beta_1=0.5)
    gan_model.compile(
        loss=[
            tf.keras.losses.MeanAbsoluteError(),   # image MAE
            tf.keras.losses.BinaryCrossentropy(),  # GAN BCE
            tf.keras.losses.MeanAbsoluteError()    # feature MAE
        ],
        loss_weights=[LAMBDA_I, LAMBDA_BCE, LAMBDA_F],
        optimizer=opt
    )
    return gan_model


```

### <font color="orange"> (2) Training the Squiggle Model

Load training data

```python
train_data = hdf5storage.loadmat('realdataset_train.mat')

mm_all   = train_data['train_wavedata']    # (N,32,64,96)
skip_all = train_data['train_skipdata']    # (N,32,64,8)
img_all  = train_data['train_imagedata']   # (N,128,256)
feat_all = train_data['train_feature']     # (N,5)

N = mm_all.shape[0]
print("N =", N)
print("mm_all:",   mm_all.shape)
print("skip_all:", skip_all.shape)
print("img_all:",  img_all.shape)
print("feat_all:", feat_all.shape)
```

Batch sampler

```python
def sample_batch(batch_size):
    idx = np.random.randint(0, N, size=batch_size)
    mm   = mm_all[idx][..., None]         # add channel -> (B,32,64,96,1)
    skip = skip_all[idx]                  # (B,32,64,8)
    img  = img_all[idx][..., None]        # (B,128,256,1)
    feat = feat_all[idx].astype('float32')
    # convert angles to radians like original code
    feat[:, 2:5] = 0.01745 * feat[:, 2:5]
    return mm.astype('float32'), skip.astype('float32'), img.astype('float32'), feat
```

Build models & training loop

```python
g_model = Generator()
d_model = Discriminator()
m_model = MetricsComputation()
gan_model = build_gan(g_model, d_model, m_model)

g_model.summary()
```

```python
from tqdm.notebook import tqdm
import os, glob, re
import numpy as np

# --- hyperparams (you can tweak) ---
BATCH_SIZE   = 4
N_EPOCHS     = 20        # total desired epochs
LR_D         = 2e-4
LR_GAN       = 1e-4
LAMBDA_I     = 1000.0    # image MAE weight
LAMBDA_BCE   = 1.0       # GAN BCE weight
LAMBDA_F     = 100.0     # feature MAE weight

steps_per_epoch = N // BATCH_SIZE
mae = tf.keras.losses.MeanAbsoluteError()
bce = tf.keras.losses.BinaryCrossentropy()

# ---------- checkpoint setup ----------
ckpt_dir = "models"
os.makedirs(ckpt_dir, exist_ok=True)

start_epoch = 1
gen_ckpts = sorted(glob.glob(os.path.join(ckpt_dir, "generator_epoch_*.h5")))

if gen_ckpts:
    last_gen = gen_ckpts[-1]
    m = re.search(r"epoch_(\d+)", os.path.basename(last_gen))
    if m:
        last_epoch = int(m.group(1))
        print(f"Resuming from epoch {last_epoch} checkpoints...")
        g_model.load_weights(os.path.join(ckpt_dir, f"generator_epoch_{last_epoch:03d}.h5"))
        d_model.load_weights(os.path.join(ckpt_dir, f"discriminator_epoch_{last_epoch:03d}.h5"))
        m_model.load_weights(os.path.join(ckpt_dir, f"metrics_epoch_{last_epoch:03d}.h5"))
        gan_model = build_gan(g_model, d_model, m_model)
        start_epoch = last_epoch + 1
else:
    print("No existing checkpoints, starting from scratch.")

# ---------- training loop ----------
for epoch in range(start_epoch, N_EPOCHS + 1):
    d_losses = []
    g_losses = []

    pbar = tqdm(range(steps_per_epoch),
                desc=f"Epoch {epoch}/{N_EPOCHS}",
                leave=False)

    for step in pbar:
        mm, skip, img, feat = sample_batch(BATCH_SIZE)

        # --- train Discriminator ---
        fake_img = g_model.predict([mm, skip], verbose=0)
        y_real = np.ones((BATCH_SIZE, 1), dtype='float32') * 0.8
        y_fake = np.zeros((BATCH_SIZE, 1), dtype='float32') + 0.2

        d_model.trainable = True
        d_loss_real = d_model.train_on_batch([mm, img],  y_real)[0]
        d_loss_fake = d_model.train_on_batch([mm, fake_img], y_fake)[0]
        d_model.trainable = False
        d_loss = 0.5 * (d_loss_real + d_loss_fake)
        d_losses.append(d_loss)

        # --- train Generator + Metrics via GAN model ---
        img_rgb = np.concatenate([img, img, img], axis=-1)
        g_loss = gan_model.train_on_batch([mm, skip], [img_rgb, y_real, feat])
        g_loss_scalar = g_loss if np.isscalar(g_loss) else g_loss[0]
        g_losses.append(g_loss_scalar)

        # always show latest averages on the bar
        pbar.set_postfix({
            "D": f"{np.mean(d_losses):.4f}",
            "G": f"{np.mean(g_losses):.4f}"
        })

    print(f"Epoch {epoch}/{N_EPOCHS} | "
          f"D loss: {np.mean(d_losses):.4f} | "
          f"G loss: {np.mean(g_losses):.4f}")

    # -------- save checkpoints for this epoch --------
    g_path = os.path.join(ckpt_dir, f"generator_epoch_{epoch:03d}.h5")
    d_path = os.path.join(ckpt_dir, f"discriminator_epoch_{epoch:03d}.h5")
    m_path = os.path.join(ckpt_dir, f"metrics_epoch_{epoch:03d}.h5")

    g_model.save(g_path)
    d_model.save(d_path)
    m_model.save(m_path)
    print(f"Saved checkpoints for epoch {epoch} to {ckpt_dir}/")

print("Training complete.")
```
